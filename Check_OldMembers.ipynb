```python
from google.colab import auth
import gspread
from google.auth import default
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

```python
import pandas as pd
path = input("Vui lòng nhập link current members: ") #danh sách hiện tại
wb = gc.open_by_url(path)
name_ws = input("Input name of the worksheet: ")
ws = wb.worksheet(name_ws)
rows = ws.get_all_values()
df = pd.DataFrame(rows)
df.columns = df.iloc[0]
df = df.iloc[1:]
print()

path_2 = input("Vui lòng nhập link data from facebook: ") #data_from_facebook
wb_2 = gc.open_by_url(path_2)
name_ws = input("Input name of the worksheet: ")
ws_2 = wb_2.worksheet(name_ws)
rows_2 = ws_2.get_all_values()
df_2 = pd.DataFrame(rows_2)
df_2.loc[-1] = ['Họ và Tên']
df_2.index = df_2.index + 1
df_2.sort_index(inplace=True)
df_2.columns = df_2.iloc[0]
df_2 = df_2.iloc[1:]
df_2 = df_2.drop(df_2[df_2["Họ và Tên"] == "Tìm hiểu thêm"].index)
print()

name_ws = input("Input name of the worksheet the input unique: ")
ws_3 = wb_2.worksheet(name_ws)
new_data = df_2["Họ và Tên"].unique()
new_data = pd.DataFrame(new_data)
new_data.loc[-1] = ['Họ và Tên']
new_data.index = new_data.index + 1
new_data.sort_index(inplace=True)
new_data.columns = new_data.iloc[0]
new_data = new_data.iloc[1:]
ws_3.update('A1', new_data.values.tolist())
```

```python
def convert(name):
      name = str(name)

      new_name = name.strip()
      rfind = new_name.rfind(" ")
      first_name = new_name[rfind+1::]

      return first_name

df["Họ và Tên"] = df['Họ và Tên'].apply(convert)
new_data['Họ và Tên'] = new_data["Họ và Tên"].apply(convert)
df_3 = pd.merge(new_data, df, how = 'left', on = 'Họ và Tên')
df_3 = df_3[df_3.isna().any(axis=1)]
df_3["Họ và Tên"]
```
