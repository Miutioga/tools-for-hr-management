```python
from google.colab import auth
import gspread
from google.auth import default
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

```python
import pandas as pd
path = input("Vui lòng nhập link file bạn muốn kiểm tra: ") #old_version
wb = gc.open_by_url(path)
name_ws = input("Input name of the worksheet")
ws = wb.worksheet(name_ws)
rows = ws.get_all_values()
df = pd.DataFrame(rows)
df.columns = df.iloc[0]
df = df.iloc[1:]
```

```python
path_2 = input("Vui lòng nhập link file bạn muốn update: ") #new_version
wb_2 = gc.open_by_url(path_2)
name_ws = input("Input name of the worksheet")
ws_2 = wb_2.worksheet(name_ws)
rows_2 = ws_2.get_all_values()
df_2 = pd.DataFrame(rows_2)
df_2.columns = df_2.iloc[0]
df_2 = df_2.iloc[1:]
```

```python
df_3 = pd.merge(df_2, df, how = 'left', on = 'Họ và Tên')
df_3.head()
```

```python
#Nhập từ dòng mới - đếm ở bảng trên
#Thường là 5, 6, 7, 8, 9
old_mssv = int(input("Dòng chứa 'MSSV' CŨ là dòng số mấy?: ")) - 1
old_dob = int(input("Dòng chứa 'Ngày sinh' CŨ là dòng số mấy?: ")) - 1
old_phone_number = int(input("Dòng chứa 'Số điện thoại' CŨ là dòng số mấy?: ")) - 1
old_email = int(input("Dòng chứa 'Email' CŨ là dòng số mấy?: ")) - 1
old_link_fb = int(input("Dòng chứa 'Link Facebook' CŨ là dòng số mấy?: ")) - 1
print()

#Thường là 14, 15, 16, 17, 18
mssv = int(input("Dòng chứa 'MSSV' là dòng số mấy?: ")) - 1
dob = int(input("Dòng chứa 'Ngày sinh' là dòng số mấy?: ")) - 1
phone_number = int(input("Dòng chứa 'Số điện thoại' là dòng số mấy?: ")) - 1
email = int(input("Dòng chứa 'Email' là dòng số mấy?: ")) - 1
link_fb = int(input("Dòng chứa 'Link Facebook' là dòng số mấy?: ")) - 1


df_3.iloc[:, mssv].fillna(df_3.iloc[:, old_mssv], inplace = True)
df_3.iloc[:, dob].fillna(df_3.iloc[:, old_dob], inplace = True)
df_3.iloc[:, phone_number].fillna(df_3.iloc[:, old_phone_number], inplace = True)
df_3.iloc[:, email].fillna(df_3.iloc[:, old_email], inplace = True)
df_3.iloc[:, link_fb].fillna(df_3.iloc[:, old_link_fb], inplace = True)

new_MSSV = df_3.iloc[:, mssv]
new_DOB = df_3.iloc[:, dob]
new_P = df_3.iloc[:, phone_number]
new_E = df_3.iloc[:, email]
new_FB = df_3.iloc[:, link_fb]
```

```python
idx = input("Dòng chứa MSSV mà bạn muốn thêm là dòng chữ gì?: ")
idx_MSSV = idx + '2'
new_data = pd.DataFrame(new_MSSV)
ws_2.update(idx_MSSV, new_data.values.tolist())

idx = input("Dòng chứa ngày sinh mà bạn muốn thêm là dòng chữ gì?: ")
idx_DOB = idx + '2'
new_data = pd.DataFrame(new_DOB)
ws_2.update(idx_DOB, new_data.values.tolist())

idx = input("Dòng chứa số điện thoại mà bạn muốn thêm là dòng chữ gì?: ")
idx_P = idx + '2'
new_data = pd.DataFrame(new_P)
ws_2.update(idx_P, new_data.values.tolist())

idx = input("Dòng chứa 'Email(@fpt.edu.vn)' mà bạn muốn thêm là dòng chữ gì?: ")
idx_E = idx + '2'
new_data = pd.DataFrame(new_E)
ws_2.update(idx_E, new_data.values.tolist())

idx = input("Dòng chứa Link FB mà bạn muốn thêm là dòng chữ gì?: ")
idx_FB = idx + '2'
new_data = pd.DataFrame(new_FB)
ws_2.update(idx_FB, new_data.values.tolist())
```
